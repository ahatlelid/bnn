```python
type_ = 2
from_ = 100
kind = ['total', 'data', 'data regularization', 'prior']

for idx, type_ in enumerate(range(2, 6)):
    plt.figure()
    plt.plot(rml_4_49[from_:,0].detach(), rml_4_49[from_:,type_].detach(), linewidth=4, color='dodgerblue')
    plt.title(f'RML, {kind[idx]} loss vs epoch, 10^4 data')
    plt.xlabel('100 epochs (10^3 batches)')
    plt.ylabel('loss')
    plt.grid()
    plt.legend([f'{kind[idx]} loss'])
    plt.savefig(f'../figures/plot_loss_vs_epoch/rml_4_49_loss_type_{type_}.png')

    plt.figure()
    plt.plot(rml_5_49[from_:,0].detach(), rml_5_49[from_:,type_].detach(), linewidth=4, color='darkorange')
    plt.title(f'RML, {kind[idx]} loss vs epoch, 10^5 data')
    plt.xlabel('10 epochs (10^3 batches)')
    plt.ylabel('loss')
    plt.legend([f'{kind[idx]} loss'])
    plt.grid()
    plt.savefig(f'../figures/plot_loss_vs_epoch/rml_5_49_loss_type_{type_}.png')
```
